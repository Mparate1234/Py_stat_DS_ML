```python

employee1=pd.read_csv("/content/drive/MyDrive/employee 1_csv.csv")
employee2=pd.read_csv("/content/drive/MyDrive/employee 2_csv.csv")
service=pd.read_csv("/content/drive/MyDrive/service_csv.csv")
print(employee1)
print("\n-----------------------------------------")
print(employee2)
print("\n-----------------------------------------")
print(service)
```

```python

## merge

                                   ## it add column wise only
employee1.merge(employee2)          # it is empty table cause there is no common data

```

```python

di={'Emplid':[11,12,3,4,5,6,7],'years_of_service':[10,51,3,5,63,15,20]}
service2=pd.DataFrame(di)
print(service2)
```

```python
employee1.merge(service2)         ## in this some values are common so it is printing it
```

```python
#joins --- 4 types ---1.left join,2.right join,3.inner join,4.outer join


pd.merge(employee1,service2,on="Emplid",how="left")          # it will join common + left table(emplyees1)
```

```python
print(employee1,"\n",service2)
pd.merge(employee1,service2,on="Emplid",how="right")
# it will join common + right(service2)
```

```python
pd.merge(employee1,service2,on="Emplid",how="inner")           # it will join only common 
                          # on          how                is used
```

```python
pd.merge(employee1,service2,on="Emplid",how="outer")           # it will join common + all 
```

```python
## Data Processing

Rating=pd.read_csv("/content/drive/MyDrive/rating_csv.csv")
print(Rating)
```

```python
## check for missing value                     NaN=not a number
Rating.isnull()
```

```python
## to check for perticular "ating column"
Rating.rating.isnull()
```

```python
## drop missing values
Rating.dropna()
```

```python
## filling the missing values

Rating.fillna("9")
  #this will fill "9" in missing place
```

```python
## filling the missing values


  #this will fill mean value of that column "rating" in missing place
    
Rating=Rating.fillna(Rating.rating.mean())
Rating
```

```python
## removing duplicates

Rating.duplicated()
```

```python
Rating.drop_duplicates()
```

```python
## check for anomaly values
import numpy as np
Rating1=Rating.replace(99999,np.nan)
Rating1
```
