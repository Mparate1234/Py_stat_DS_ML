```python
import pandas as pd
data=pd.read_csv('/content/drive/MyDrive/Breast_Cancer_Dataset.csv')
data.head(3)
```

### 1. Identify the null values and remove the null rows and columns by using the dropna()function

```python
data.isnull().sum()
```

### 2. Encode the 'diagnosis' column using the LabelEncoder()

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
data['diagnosis']=labelencoder.fit_transform(data['diagnosis'])
data.head()
```

### 3. Considering the 'diagnosis' column as the target, separate the target variable and the feature vectors

```python
from sklearn.model_selection import train_test_split,KFold
x=data.drop(['diagnosis'],axis=1)
y=data['diagnosis']
```

### 4. Split the dataset into the training set and test set in a 70:30 ratio


```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
x.head(3)
```

```python
y.head(3)
```

### 5. Building a Logistic Regression, Naive Bayes, Decision Tree (CART), K-NN, SVM, and RandomForestClassifier models; Also print their accuracies

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

models = [('Logistic Regression', LogisticRegression()), 
          ('Naive Bayes', GaussianNB()), 
          ('Decision Tree (CART)', DecisionTreeClassifier()), 
          ('K-NN', KNeighborsClassifier()), 
          ('SVM', SVC()), 
          ('Random Forest Classifier', RandomForestClassifier())]
```

```python
for name, model in models:
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f'{name} accuracy: {accuracy:.2f}')
```

# Problem Statement 4:
The ‘seeds.csv’ dataset contains the data about the wheat seeds, the
‘Type’ column consisit of three unique values, 1, 2, 3, which are classified based on the
charecterstics of seeds entailing in other columns.

Load the ‘seeds.csv’ dataset into a DataFrame and perform the following tasks:

```python
import pandas as pd
import numpy as np
```

```python
seeds_data=pd.read_csv('/content/drive/MyDrive/seeds.csv')
seeds_data.head()
```

### 1. Considering the 'Type' column as target, analyze the target column by printing the unique values

```python
pd.DataFrame({'Count':seeds_data.isnull().sum(),'Percentage':seeds_data.isnull().sum()/seeds_data.shape[0]})
```

```python
target_ratio=pd.DataFrame({'Counts':seeds_data.Type.value_counts(),'Percentage':seeds_data.Type.value_counts()/len(seeds_data)})
target_ratio
```

### 2. Separate the feature vectors and the target variable

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
x=seeds_data.drop(['Type'],axis=1)
y=seeds_data['Type']
```

```python
x.head(3)
```

```python
y.head(3)
```

### 3. Split the dataset into train and test sets in a 70:30 ratio


```python
X_train, X_test, y_train, y_test=train_test_split(x,y,random_state=2,test_size=0.3)
```

### 4. Build a Decision Tree Classifier and a GaussianNB model and print their accuracy scores


```python
dt_clf=DecisionTreeClassifier(max_depth=1)
dt_clf.fit(X_train,y_train)
dt_pred=dt_clf.predict(X_test)
acc_dt=round(accuracy_score(y_test,dt_pred),3)
print("Accuracy score of the Dicision tree model is :",acc_dt)
```

```python
gnb_model=GaussianNB()
gnb_model.fit(X_train,y_train)
gnb_pred=gnb_model.predict(X_test)
acc_gnb=round(accuracy_score(y_test,gnb_pred),3)
print("Accuracy score of the GaussianNB model is :",acc_gnb)
```

```python
ada=AdaBoostClassifier(n_estimators=100,learning_rate=0.6)
ada.fit(X_train,y_train)
ada_pred=ada.predict(X_test)
acc_ada=round(accuracy_score(y_test,ada_pred),3)
print(acc_ada)
```

### 5. For the Decision Tree Classifier and a GaussianNB models boost the accuracy using ADA Boost Classifier and compare the accuracy scores with original models using a bar plot

```python
ada_gnb=AdaBoostClassifier(base_estimator=GaussianNB(),n_estimators=70,learning_rate=0.5)
ada_gnb.fit(X_train,y_train)
agnb_pred=ada_gnb.predict(X_test)
acc_agnb=round(accuracy_score(y_test,agnb_pred),3)
print(acc_agnb)
```

```python
plt.figure(figsize=(10,2))
plt.barh(np.arange(4),[acc_dt,acc_gnb,acc_ada,acc_agnb],\
tick_label=['Decision Tree','GaussianNB','AdaBoost','AdaBoost(GaussianNB)'])
```
